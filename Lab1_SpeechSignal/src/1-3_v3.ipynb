```python
#录制音频
def record_audio(duration):
    import pyaudio
    import wave

    CHUNK = 1024
    FORMAT = pyaudio.paInt16
    CHANNELS = 1
    RATE = 44100
    WAVE_OUTPUT_FILENAME = "demo.wav"

    p = pyaudio.PyAudio()

    stream = p.open(format=FORMAT,
                    channels=CHANNELS,
                    rate=RATE,
                    input=True,
                    frames_per_buffer=CHUNK)

    print("* 录音中...")

    frames = []

    for i in range(0, int(RATE / CHUNK * duration)):
        data = stream.read(CHUNK)
        frames.append(data)

    print("* 录音结束!")

    stream.stop_stream()
    stream.close()
    p.terminate()

    wf = wave.open(WAVE_OUTPUT_FILENAME, 'wb')
    wf.setnchannels(CHANNELS)
    wf.setsampwidth(p.get_sample_size(FORMAT))
    wf.setframerate(RATE)
    wf.writeframes(b''.join(frames))
    wf.close()
    
record_audio(duration=5) #录音5秒,并保存为demo.wav
```

```python
import numpy as np

def pitch_Corr(x, wnd, inc, T1, fs, miniL=10):
    """
    自相关法基音周期检测函数
    :param x: 语音信号
    :param wnd: 窗函数或窗长
    :param inc: 帧移
    :param T1: 门限
    :param fs: 采样率
    :param miniL: 语音段的最小帧数
    :return voiceseg, vsl, SF, Ef, period: 语音段的起始点和终止点，语音段数，语音段标记，非语音段标记，基音周期
    """
    y = enframe(x, wnd, inc)  # 分帧
    fn = y.shape[0]  # 帧数
    voiceseg, vsl, SF, Ef = pitch_vad(x, wnd, inc, T1, miniL)  # 语音分段
    if not voiceseg:  # 如果voiceseg为空
        print("No voice segments detected.")
        return None, None, None, None, None
    lmax = fs // 60  # 基音周期的最大值
   

    period = np.zeros(fn)
    for i in range(vsl):  # 在所有语音段中
        ixb = voiceseg[i]['start']  # 语音段的起始帧
        ixd = voiceseg[i]['duration']  # 语音段的帧数
        for k in range(ixb, ixb + ixd):
            frame = y[k, :]
            ru = np.correlate(frame, frame, mode='full')  # 计算短时自相关
            ru = ru[len(frame)-1:]  # 仅保留正的延迟部分
            ru[:lmin] = 0  # 忽略小于最小基音周期的延迟
            max_pos = np.argmax(ru[:lmax])  # 在允许的范围内找到最大值位置
            period[k] = max_pos + lmin  # 计算并存储基音周期的估值，加上lmin是因为索引从0开始，而基音周期的实际计算应从lmin开始
    return voiceseg, vsl, SF, Ef, period
```

实验1-3：MFCC提取
需要补充下列方法: Nmfcc

```python
import numpy as np
from scipy.signal import lfilter, hamming
from scipy.fftpack import dct

def preemphasis(signal, coeff=0.97):
    """
    对信号进行预加重。
    :param signal: 输入信号。
    :param coeff: 预加重系数。
    :return: 预加重后的信号。
    """
    return np.append(signal[0], signal[1:] - coeff * signal[:-1])

def enframe(signal, frame_length, frame_step):
    """
    将信号分帧。
    :param signal: 输入信号。
    :param frame_length: 帧长。
    :param frame_step: 帧移。
    :return: 分帧后的信号矩阵。
    """
    signal_length = len(signal)
    num_frames = int(np.ceil(float(np.abs(signal_length - frame_length)) / frame_step))
    pad_signal_length = num_frames * frame_step + frame_length
    z = np.zeros((pad_signal_length - signal_length))
    pad_signal = np.append(signal, z)
    indices = np.tile(np.arange(0, frame_length), (num_frames, 1)) + np.tile(np.arange(0, num_frames * frame_step, frame_step), (frame_length, 1)).T
    frames = pad_signal[indices.astype(np.int32, copy=False)]
    return frames

def mel_filters(fs, nfft, n_filters=40):
    """
    创建Mel滤波器组。
    :param fs: 采样率。
    :param nfft: FFT点数。
    :param n_filters: 滤波器组的数量。
    :return: Mel滤波器组。
    """
    low_freq_mel = 0
    high_freq_mel = (2595 * np.log10(1 + (fs / 2) / 700))  # 将Hz转换为Mel
    mel_points = np.linspace(low_freq_mel, high_freq_mel, n_filters + 2)  # 均匀分布的Mel频率
    hz_points = (700 * (10**(mel_points / 2595) - 1))  # 将Mel转换回Hz
    bin = np.floor((nfft + 1) * hz_points / fs)

    fbank = np.zeros((n_filters, int(np.floor(nfft / 2 + 1))))
    for m in range(1, n_filters + 1):
        f_m_minus = int(bin[m - 1])   # 左
        f_m = int(bin[m])             # 中心
        f_m_plus = int(bin[m + 1])    # 右

        for k in range(f_m_minus, f_m):
            fbank[m - 1, k] = (k - bin[m - 1]) / (bin[m] - bin[m - 1])
        for k in range(f_m, f_m_plus):
            fbank[m - 1, k] = (bin[m + 1] - k) / (bin[m + 1] - bin[m])

    return fbank


def Nmfcc(x, fs, p, frameSize, inc, nfft=512, n_dct=12):
    """
    计算mfcc系数。
    :param x: 输入信号。
    :param fs: 采样率。
    :param p: Mel滤波器组的个数。
    :param frameSize: 分帧的每帧长度。
    :param inc: 帧移。
    :return: mfcc系数。
    """
    # 预加重处理
    x_preemphasized = preemphasis(x)

    # 分帧
    frames = enframe(x_preemphasized, frameSize, inc)

    # 加窗
    frames *= np.hamming(frameSize)

    # 计算FFT
    fft_result = np.fft.rfft(frames, n=nfft)

    # 计算能量谱
    power_spectrum = np.abs(fft_result) ** 2
    
    # 计算通过Mel滤波器的能量
    # 此处应调用mel_filters或类似函数生成Mel滤波器组，由于代码略去，需要补充
    bank = mel_filters(fs, nfft, p)
    mel_energy = np.dot(power_spectrum, bank.T)
    
    # 计算对数能量
    log_mel_energy = np.log(mel_energy + np.finfo(float).eps)
    
    # 计算DCT倒谱
    mfcc = dct(log_mel_energy, type=2, axis=-1, norm='ortho')[:, :n_dct]

    return mfcc
```
